**Read the CSV File into a Pandas DataFrame**

Load the Stack Overflow survey data into a Pandas DataFrame:


```python
# Read the CSV file
df = pd.read_csv('survey-data.csv')

# Display the first few rows of the data
df.head()
```

**Create a SQLite Database and Insert the Data**

Now, let's create a new SQLite database (`survey-data.sqlite`) and insert the data from the DataFrame into a table using the sqlite3 library:


```python
import sqlite3

# Create a connection to the SQLite database
conn = sqlite3.connect('survey-data.sqlite')

# Write the dataframe to the SQLite database
df.to_sql('main', conn, if_exists='replace', index=False)


# Close the connection
conn.close()
```

**Verify the Data in the SQLite Database**
Verify that the data has been correctly inserted into the SQLite database by running a simple query:


```python
# Reconnect to the SQLite database
conn = sqlite3.connect('survey-data.sqlite')

# Run a simple query to check the data
QUERY = "SELECT * FROM main LIMIT 5"
df_check = pd.read_sql_query(QUERY, conn)

# Display the results
print(df_check)
```

## Hands-on Lab


### Visualizing the Distribution of Data

**Histograms**

Plot a histogram of CompTotal (Total Compensation).


```python
## Write your code here
col = pd.read_sql_query("SELECT CompTotal from main", conn)['CompTotal'].dropna()
col = col[col <col.quantile(0.99)]

plt.figure(figsize=(10, 5))
plt.hist(col, bins=50, edgecolor='black')
plt.title('Distribution of Total Compensation (CompTotal)')
plt.xlabel('CompTotal')
plt.ylabel('Frequency')
plt.show()
```

**Box Plots**

Plot a box plot of Age.


```python
## Write your code here
age_map = {'Under 18 years old': 16, '18-24 years old': 21, '25-34 years old': 30,
           '35-44 years old': 40, '45-54 years old': 50, '55-64 years old': 60,
           '65 years or older': 70, 'Prefer not to say': None}
ages = pd.read_sql_query('SELECT Age FROM main',conn)['Age'].map(age_map).dropna()

plt.figure(figsize=(4,3))
plt.boxplot(ages)
plt.title('Box plot of Respondent Ages')
plt.ylabel('Age Rank')

plt.show()
```

### Visualizing Relationships in Data

**Scatter Plots**

Create a scatter plot of Age and WorkExp.


```python
## Write your code here
age_map = {'Under 18 years old': 16, '18-24 years old': 21, '25-34 years old': 30,
           '35-44 years old': 40, '45-54 years old': 50, '55-64 years old': 60,
           '65 years or older': 70, 'Prefer not to say': None}
data = pd.read_sql_query("SELECT Age, WorkExp FROM main", conn)
data['Age_num'] = data['Age'].map(age_map)


plt.figure(figsize=(8,5))
plt.scatter(data['Age_num'],data['WorkExp'], alpha=0.4)
plt.title('Age vs Work Experience')
plt.xlabel('Age')
plt.ylabel('WorkExp (years)')
plt.show()
```

**Bubble Plots**

Create a bubble plot of `TimeSearching` and `Frustration` using the Age column as the bubble size.


```python
#pd.read_sql_query('SELECT Frustration FROM main', conn).value_counts().head(10)
```

```python
## Write your code here
time_map = {'Less than 15 minutes a day': 7.5, '15-30 minutes a day': 22.5,
            '30-60 minutes a day': 45, '60-120 minutes a day': 90,
            'Over 120 minutes a day': 150}
data = pd.read_sql_query('SELECT TimeSearching , Frustration, Age FROM main',conn)
data['TimeSearching_min'] = data['TimeSearching'].map(time_map)
data['Frustration_count'] = data['Frustration'].str.split(';').str.len()
data['Age_num'] = data['Age'].map(age_map)
data = data.dropna(subset=['TimeSearching_min','Frustration_count','Age_num'])

plt.figure(figsize=(7,4))
plt.scatter(data['TimeSearching_min'],data['Frustration_count'],s=data['Age_num'],alpha =0.3)
plt.title('Time Searching vs Number of Frustrations (bubble size = Age)')
plt.xlabel('Time searching (minutes per day)')
plt.ylabel('Number of frustrations reported')
plt.show() 
```

### Visualizing Composition of Data

**Pie Charts**

Create a pie chart of the top 5 databases(`DatabaseWantToWorkWith`) that respondents wish to learn next year.


```python
## Write your code here
dbs = pd.read_sql_query("SELECT DatabaseWantToWorkWith FROM main", conn)['DatabaseWantToWorkWith']
top5 = dbs.dropna().str.split(';').explode().str.strip().value_counts().head(5)

plt.figure(figsize=(5, 5))
plt.pie(top5, labels=top5.index, autopct='%1.1f%%', startangle=90)
plt.title('Top 5 Databases Respondents Want to Work With')
plt.show()
```

**Stacked Charts** 

Create a stacked bar chart of median `TimeSearching` and `TimeAnswering` for the age group 30 to 35.


```python
## Write your code here
data = pd.read_sql_query("SELECT Age, TimeSearching, TimeAnswering FROM main", conn)
data['Age_numeric'] = data['Age'].map(age_map)
data['TimeSearching_min'] = data['TimeSearching'].map(time_map)
data['TimeAnswering_min'] = data['TimeAnswering'].map(time_map)
group = data[(data['Age_numeric'] >= 30) & (data['Age_numeric'] <= 35)]
medians = group[['TimeSearching_min', 'TimeAnswering_min']].median()

plt.figure(figsize=(5, 6))
plt.bar('Ages 30-35', medians['TimeSearching_min'], label='Time Searching')
plt.bar('Ages 30-35', medians['TimeAnswering_min'], bottom=medians['TimeSearching_min'], label='Time Answering')
plt.title('Median Daily Minutes: Searching vs Answering (Ages 30-35)')
plt.ylabel('Minutes per day (median)')
plt.legend()
plt.show()
```

### Visualizing Comparison of Data

**Line Chart**

Plot the median `CompTotal` for all ages from 45 to 60.


```python
## Write your code here
data = pd.read_sql_query("SELECT Age, CompTotal FROM main", conn)
data['Age_numeric'] = data['Age'].map(age_map)
subset = data[(data['Age_numeric'] >= 45) & (data['Age_numeric'] <= 60)]
median_comp = subset.groupby('Age_numeric')['CompTotal'].median()

plt.figure(figsize=(6, 4))
plt.plot(median_comp.index, median_comp.values, marker='o')
plt.title('Median Total Compensation for Ages 45-60')
plt.xlabel('Age')
plt.ylabel('Median CompTotal')
plt.show()
```
